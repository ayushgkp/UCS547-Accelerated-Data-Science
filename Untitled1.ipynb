1. Reverse the NumPy array: arr = np.array([1, 2, 3, 6, 4, 5]) )

```python
import numpy as np

arr = np.array([1, 2, 3, 6, 4, 5])
rev = arr[::-1]
print(rev)
```

2. Compare the following numpy arrays: arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])

```python
import numpy as np

arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])

print(arr1 == arr2)
print(np.array_equal(arr1, arr2))
```

3. Find the most frequent value and their indice(s) in the following arrays:


i. x = np.array([1,2,3,4,5,1,2,1,1,1])

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])

values, counts = np.unique(x, return_counts=True)
most_freq = values[np.argmax(counts)]
indices = np.where(x == most_freq)

print("Most frequent value:", most_freq)
print("Indices:", indices)
```

ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
y = np.array([1,1,1,2,3,4,2,4,3,3])

values, counts = np.unique(y, return_counts=True)
most_freq = values[np.argmax(counts)]
indices = np.where(y == most_freq)

print("Most frequent value:", most_freq)
print("Indices:", indices)
```

4. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),
find:

 i.Sum of all elements

```python
import numpy as np

gfg = np.matrix([[4, 1, 9],
                 [12, 3, 1],
                 [4, 5, 6]])

total_sum = gfg.sum()
print("Sum of all elements:", total_sum)
```

ii. Sum of all element’s row-wise

```python
import numpy as np

gfg = np.matrix([[4, 1, 9],
                 [12, 3, 1],
                 [4, 5, 6]])

row_sum = gfg.sum(axis=1)
print("Row-wise sum:")
print(row_sum)
```

iii. Sum of all element’s column-wise

```python
import numpy as np

gfg = np.matrix([[4, 1, 9],
                 [12, 3, 1],
                 [4, 5, 6]])

col_sum = gfg.sum(axis=0)
print("Column-wise sum:")
print(col_sum)
```
